```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns
sns.set()
```

```python
data = pd.read_csv('Satscores.csv')
```

```python
data
```

```python
data.describe()
```

```python
# y = b0+b1x1
y = data['GPA']
x1 = data['SAT']
```

```python
plt.scatter(x1,y)
plt.xlabel('SAT',fontsize=20)
plt.ylabel('GAT',fontsize=20)
plt.show()
```

```python
x = sm.add_constant(x1)
results = sm.OLS(y,x).fit()
results.summary()
```

```python
plt.scatter(x1,y)
yhat = -0.002*x1 + 3.3684
fig = plt.plot(x1,yhat, lw=4, c='orange', label='regression line')
plt.xlabel('SAT',fontsize=20)
plt.ylabel('GAT',fontsize=20)
plt.show()
```
